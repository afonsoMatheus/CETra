```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = {}
with open("/home/afonso/Documentos/Dissertação/Masters Degree/tests/stress_test/output/graph", 'r') as file:
    data['Monic'] = file.readline().split(" ")[:-1]
    data['CETra'] = file.readline().split(" ")[:-1]
    data['Number'] = file.readline().split(" ")[:-1]
df = pd.DataFrame(data)
```

```python
df['Monic'] = pd.to_numeric(df['Monic'])
df['CETra'] = pd.to_numeric(df['CETra'])
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(data = df, x = 'Number', y = 'CETra', marker = "o", markersize=5, legend='auto' , label='CETra')
sns.lineplot(data = df, x = 'Number', y = 'Monic', marker = "o", markersize=5, legend='auto', label='Monic')

#y_ticks = np.arange(0, 49, 2)
#plt.yticks(y_ticks)
plt.grid(axis = 'y')

plt.xlabel('Number of Sensors')
plt.ylabel('Time (Microseconds)')
plt.savefig('saving-a-high-resolution-seaborn-plot.png', dpi=300)
```
