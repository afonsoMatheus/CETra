```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = {}
with open("/home/afonso/Documentos/Dissertação/Masters Degree/tests/cont_test/graph", 'r') as file:
    data['C1'] = file.readline().split(" ")[:-1]
    data['C2'] = file.readline().split(" ")[:-1]
    data['C3'] = file.readline().split(" ")[:-1]
    data['Time'] = file.readline().split(" ")[:-1]
df = pd.DataFrame(data)
```

```python
df["C1"] = pd.to_numeric(df["C1"])
df["C2"] = pd.to_numeric(df["C2"])
df["C3"] = pd.to_numeric(df["C3"])
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(data = df, x = 'Time', y = "C1", marker = "o", markersize=5, legend='auto', label='C1')
sns.lineplot(data = df, x = 'Time', y = "C2", marker = "o", markersize=5, legend='auto' , label='C2')
sns.lineplot(data = df, x = 'Time', y = "C3", marker = "o", markersize=5, legend='auto' , label='C3')

y_ticks = np.arange(0, 49, 2)
plt.yticks(y_ticks)
plt.grid(axis = 'y')

plt.xlabel('Time')
plt.ylabel('Internal Statistic')
plt.savefig('saving-a-high-resolution-seaborn-plot.png', dpi=300)
```
